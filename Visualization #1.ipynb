## Visualization #1 (No Time)
    Do higher GPA Avgs correlate with higher enrollment per institution?

    GPA and Enrollment Data by Institution

```python
import pandas as pd
import bokeh as bk
import seaborn as sb
%matplotlib inline
from bokeh.plotting import *
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.models.glyphs import Line as Line_glyph, Oval
from bokeh.models import Range1d, LinearAxis, FactorRange, LabelSet
from bokeh.charts import Bar, Scatter, BoxPlot
from bokeh.charts.attributes import cat
from bokeh.layouts import gridplot
import numpy as np
pd.options.display.max_rows = 50
pd.options.display.max_columns = 999
```

```python
#Reading In Data, Filling NA
ncwit_df = pd.read_csv("NCWIT-Data Cleaned.csv", low_memory = False)
ncwit_df = ncwit_df.fillna(0)
ncwit_df.head()
```

```python
#Grouping by Institution, Avg GPA
gpaAvg_df = ncwit_df[['Institution', 'Totals, Female: Cumulative GPA (Tot. F)', 'Totals, Male: Cumulative GPA (Tot. M)']].copy()
gpaAvg_df = gpaAvg_df.groupby(['Institution'], as_index = False).mean()
gpaAvg_df
```

```python
#Grouping By Institution
totenrl_df = ncwit_df[['Institution', 'School Year', 'Enroll, Female: Asian (Enrl F)', 'Enroll, Female: Black/African American (Enrl F)', 'Enroll, Female: Black/African American (Enrl F)', 'Enroll, Female: Hispanics of any race (Enrl F)', 'Enroll, Female: American Indian/Alaska Native (Enrl F)', 'Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)', 'Enroll, Female: Two or more races (Enrl F)', 'Enroll, Female: White (Enrl F)']].copy()
totenrl_df = totenrl_df.groupby(['Institution'], as_index = False).sum()
totenrl_df 
```

```python
#Summing Female Enrollment over all years
totenrl_df['Total Female Enrollment'] = totenrl_df.sum(axis = 1).subtract(totenrl_df['Institution'])
totenrl_df
```

```python
#Merging the Two DFs
viz_df = pd.merge(totenrl_df, gpaAvg_df, on = 'Institution')

#Renaming Annoying Column
viz_df.rename(columns={'Totals, Female: Cumulative GPA (Tot. F)':'Average_Female_GPA'}, inplace=True)
viz_df.rename(columns={'Total Female Enrollment':'Total_Female_Enrollment'}, inplace=True)

#Sorting Values based off Enrollment
viz_df = viz_df.sort_values('Total_Female_Enrollment', ascending = False)

#Reducing DF to exclude those who do not have enrollment numbers
viz_df = viz_df[viz_df['Total_Female_Enrollment'] != 0]

#Selecting Data for Figure 1
viz_df = viz_df.loc[:, ('Institution', 'Total_Female_Enrollment', 'Average_Female_GPA')]
viz_df
```

```python
#Creating Output File
output_file("Visualization #1")

#Creating Source Data Obj from df
source = ColumnDataSource(data = viz_df)

#Interactivity
hover = HoverTool(tooltips=[
    ("Institution", "@x"),
    ("Total Female Enrollment", "@y"),
])
hover1 = HoverTool(tooltips=[
    ("Institution", "@x"),
    ("GPA", "@y"),
])

#Creating High Level Bar Chart
s1 = Bar(viz_df,values='Total_Female_Enrollment',
             label=cat(columns='Institution',sort=False), plot_height = 600,
             plot_width=600, legend = False, tools = [hover],
            title = "Sum of Female Enrollment by Institution")

s3 = Scatter(viz_df,x='Institution',
            y = 'Average_Female_GPA', plot_height = 600,
             plot_width=600, legend = False, tools = [hover1],
            title = "Average GPA by Institution", color = "blue")

s3.xaxis.axis_label = "Institution"
s3.yaxis.axis_label = "Average GPA by Institution"

#Adding Regression Line
regression = np.polyfit(viz_df['Institution'], viz_df['Average_Female_GPA'], 1)
r_x, r_y = zip(*((i, i*regression[0] + regression[1]) for i in range(500)))
s3.line(r_x, r_y, color="red")

#s2.x_range = FactorRange(factors=viz_df['Institution'].tolist())

p = gridplot([[s1, s3]], toolbar_location=None)

show(p)
```
