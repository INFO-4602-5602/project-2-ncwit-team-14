## Visualization #1 (No Time)
    Do higher GPA Avgs correlate with higher enrollment per institution?

    GPA and Enrollment Data by Institution

```python
import pandas as pd
import bokeh as bk
from bokeh.plotting import figure, output_file, show, output_notebook
from bokeh.models import ColumnDataSource
from bokeh.models.glyphs import Line as Line_glyph
from bokeh.models import Range1d, LinearAxis, FactorRange
import numpy as np
pd.options.display.max_rows = 50
pd.options.display.max_columns = 999
```

```python
#Reading In Data, Filling NA
ncwit_df = pd.read_csv("NCWIT-Data Cleaned.csv", low_memory = False)
ncwit_df = ncwit_df.fillna(0)
ncwit_df.head()
```

```python
#Grouping by Institution, Avg GPA
gpaAvg_df = ncwit_df[['Institution', 'Totals, Female: Cumulative GPA (Tot. F)', 'Totals, Male: Cumulative GPA (Tot. M)']].copy()
gpaAvg_df = gpaAvg_df.groupby(['Institution'], as_index = False).mean()
gpaAvg_df
```

```python
#Grouping By Institution
totenrl_df = ncwit_df[['Institution', 'School Year', 'Enroll, Female: Asian (Enrl F)', 'Enroll, Female: Black/African American (Enrl F)', 'Enroll, Female: Black/African American (Enrl F)', 'Enroll, Female: Hispanics of any race (Enrl F)', 'Enroll, Female: American Indian/Alaska Native (Enrl F)', 'Enroll, Female: Native Hawaiian/Other Pacific Islander (Enrl F)', 'Enroll, Female: Two or more races (Enrl F)', 'Enroll, Female: White (Enrl F)']].copy()
totenrl_df = totenrl_df.groupby(['Institution'], as_index = False).sum()
totenrl_df 
```

```python
#Summing Female Enrollment over all years
totenrl_df['Total Female Enrollment'] = totenrl_df.sum(axis = 1).subtract(totenrl_df['Institution'])
totenrl_df
```

```python
#Merging the Two DFs
viz_df = pd.merge(totenrl_df, gpaAvg_df, on = 'Institution')

#Sorting Values based off Enrollment
viz_df = viz_df.sort_values('Total Female Enrollment', ascending = False)

#Reducing DF to exclude those who do not have enrollment numbers
viz_df = viz_df[viz_df['Total Female Enrollment'] != 0]
viz_df.head(10)
```

```python
#Creating Output File
output_notebook()

#Setting Dimensions
p = figure(plot_width = 800, plot_height = 400)

#Creating Point Renderer
p.cross(viz_df.index,viz_df['Totals, Female: Cumulative GPA (Tot. F)'].values,line_width = 0.5, color = 'red', y_range_name = 'GPA')

#Creating alterante range for GPA
p.extra_y_ranges = {'GPA': Range1d(start = 0, end = 4.0)}
p.add_layout(LinearAxis(y_range_name = 'GPA'), 'right')

#Creating Bar Values
h = viz_df['Total Female Enrollment'].values

#Adding Bar Renderer
p.vbar(x = viz_df.index, bottom = 0, top = h, width = .5)




show(p)
```
